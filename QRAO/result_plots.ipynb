```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mpl
from matplotlib import animation
```

```python
file_path = 'graph_3reg_1_results.txt'

# Load the data
data = np.loadtxt(file_path)  # Skip the first row with column names

# Print the loaded data
print(data)
N_vertex=[]
N_qubits=[]
gw=[]
dgw=[]
sep=[]
dsep=[]
ent=[]
dent=[]
for res in data:
    N_vertex.append(res[0])
    N_qubits.append(res[1])
    gw.append(res[2])
    dgw.append(res[3])
    sep.append(res[4])
    dsep.append(res[5])
    ent.append(res[6])
    dent.append(res[7])
plt.errorbar(N_vertex,gw,dgw,label='GW',marker='*')
plt.errorbar(N_vertex,sep,dsep,label='sep',marker='*')
plt.errorbar(N_vertex,ent,dent,label='ent',marker='*')
plt.xlabel('Number of vertices')
plt.ylabel('Approximation ratio')
plt.ylim(0,1.1)
plt.title('3-regular graph')
plt.legend()
plt.show()
```
